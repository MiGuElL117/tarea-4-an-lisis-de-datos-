```python
from sklearn.cluster import KMeans
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
```

```python
data = {
    'CustomerID': [1, 2, 3, 4, 5],
    'Gender': ['Male', 'Male', 'Female', 'Female', 'Female'],
    'Age': [19, 21, 20, 23, 31],
    'Annual Income (k$)': [15, 15, 16, 16, 17],
    'Spending Score (1-100)': [39, 81, 6, 77, 40]
}

datos = pd.DataFrame(data)
```

```python
X = datos[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
gender_encoder = OneHotEncoder()
gender_encoded = gender_encoder.fit_transform(datos[['Gender']]).toarray()
X = pd.concat([X, pd.DataFrame(gender_encoded, columns=gender_encoder.get_feature_names_out(['Gender']))], axis=1)
```

```python
param_grid = {'n_clusters': [2, 3, 4, 5, 6]}
```

```python
modelo = KMeans()
```

```python
grid_search = GridSearchCV(estimator=modelo, param_grid=param_grid, cv=5)
```

```python
grid_search.fit(X)
```

```python
mejor_modelo = grid_search.best_estimator_
```

```python
y_pred = mejor_modelo.predict(X)
```

```python
print("Etiquetas de los clústeres:", y_pred)
```

```python
#punto 5
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_avg = silhouette_score(X, y_pred)
```

```python
print("Índice de silueta:", silhouette_avg)
```

```python
from sklearn.metrics import calinski_harabasz_score
```

```python
calinski_harabasz_score = calinski_harabasz_score(X, y_pred)
```

```python
print("Índice de Calinski-Harabasz:", calinski_harabasz_score)
```

```python
from sklearn.metrics import davies_bouldin_score
```

```python
davies_bouldin_score = davies_bouldin_score(X, y_pred)
```

```python
print("Índice de Davies-Bouldin:", davies_bouldin_score)
```
